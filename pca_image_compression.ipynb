# **PCA (Dimention Reduction)**

```python
Goal="""In this code I explore image compression with Principal Component Analysis (PCA), a key technique for reducing data dimensionality in machine learning. 
I will transform images into smaller, lower-dimensional forms while keeping most of the important information."""
```

## Data Preprocessing

```python
"""Data source: https://drive.google.com/drive/folders/1GiM_pE5pkHU-OJIoyV2G9Mq4fX_0yCej"""


# first we need to transform data to a matrix with elements in [0,255]
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

#opening the pic
path = r"C:\Users\Maryam\Desktop\E2.ML\Google-trex.png"
image = Image.open(path)

#reducing the dimention of the channel in matrix.shape
image=image.convert('L')
#converting the image to an array
imagematrix = np.array(image)

print("Shape of matrix:", imagematrix.shape) #(height,width,channel(:red,green,blue,transparency))
print(imagematrix)

#showing the real pic
plt.imshow(imagematrix,cmap='gray')
plt.title("Original Image")
plt.axis("off")
plt.show()
```

## Applying PCA for Image Compression

### D=200

```python
from sklearn.decomposition import PCA

dimention=200 #dimention of the new matrix (compressed pic) (the max is 255)
pcarunner=PCA(n_components=dimention)
compress=pcarunner.fit_transform(imagematrix) #reducing the height, width & dimention
newpic= pcarunner.inverse_transform(compress) #keeping the dimention but reconstructing the height and width as before

#showing the real pic
plt.imshow(newpic,cmap='gray')
plt.title("reducted Image(D=200)")
plt.axis("off")
plt.show()
```
